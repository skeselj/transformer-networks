```python
import os, h5py
import matplotlib.pyplot as plt, matplotlib.image as mpimg
from __future__ import show
```

```python
log_dir = 'logs/'
os.listdir(log_dir)  # runs which were logged
```

```python
show('onelevel/', 2*1000, orientation='square')  # has to be multiple of 1000
```

```python
show('dump/', 10*200)
```

```python
def show(run_dir, iteration, orientation='column'):  
    path = log_dir + run_dir + str(iteration) 

    target = mpimg.imread(log_dir + run_dir + "target" + str(iteration) + '.png')
    frame = mpimg.imread(log_dir + run_dir + "frame" + str(iteration) + '.png')
    pred = mpimg.imread(log_dir + run_dir + "pred" + str(iteration) + '.png')
    field = mpimg.imread(log_dir + run_dir + "field" + str(iteration) + '.png')

    if orientation == 'square':
        fig = plt.figure()
        ax1 = fig.add_subplot(2,2,1)
        ax1.imshow(target)
        ax2 = fig.add_subplot(2,2,2)
        ax2.imshow(frame)
        ax3 = fig.add_subplot(2,2,3)
        ax3.imshow(pred)
        ax4 = fig.add_subplot(2,2,4)
        ax4.imshow(field)
        
    elif orientation == 'column':
        plt.imshow(target)
        plt.show()
        plt.imshow(frame)
        plt.show()
        plt.imshow(pred)
        plt.show()
        plt.imshow(field)
        
    plt.show()
```
