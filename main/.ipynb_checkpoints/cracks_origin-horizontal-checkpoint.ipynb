## Dataset

A pair in the dataset consists of a "frame": the cracked image, and a "target": the original image before it was cracked. 

To generate a pair, I randomly sample a target from pinky40 and then get the frame by deforming it in the following way:
 - pick an angle $\theta$ uniformly in $[0,2\pi)$
 - define line $l$ as the line through the center of the image at angle $\theta$
 - pick an integer $d$ uniformly in $[5,25]$
 - shift every pixel left of $l$ to the left by $\lfloor d/2 \rfloor$, every pixel right of $l$ to the right by $\lceil d/2 \rceil$


There are 10,000 training images, which are shuffled every epoch, and 1,000 testing images. The frames are 256x256 and the targets are 224x224; this is to ensure no empty borders.

## Model

SpyNet with 5 layers (4 downsamples). Each layer has 5 same convolutions, each with 7x7 kernels.

The loss function consists of MSE and first-order smoothness with a coefficient that makes it comparable in size to MSE. 

The main things I changed since last week are finding a better (initial LR, LR decay) pair and adding more levels to the net.

## Discussion

### Patchiness

The images where the top and bottom is patchy look bad. However, the images they generate don't look patchy. This could be because although the vectors have many different directions, but they are close to 0 in magnitude so it doesn't matter.

Let's look at the elementwise 2-norm of the vector field of a problematic example. Below I look at the 4th example from above.

```python
import numpy as np
import h5py
import matplotlib.pyplot as plt

f = h5py.File('logs/crack/origin-1/pancakes/test_160_data.h5', 'r')
x, y = f['field'][:,:,0], f['field'][:,:,1]
Ix, Iy = np.meshgrid(np.linspace(-7.0/8,7.0/8,224), np.linspace(-7.0/8,7.0/8,224))
x = x - Ix; y = y - Iy
m = np.sqrt(np.multiply(x,x) + np.multiply(y,y))
plt.imshow(np.squeeze(m[:,:]))
plt.title('Elementwise 2-norm of field from 4th example above')
plt.show()
f.close()
```

```python
patchy_avg_norm = 0.5 * (np.mean(m[:25,:]) + np.mean(m[199:,:]))
print 'average 2-norm in top and bottom 25 rows is ', patchy_avg_norm
smooth_avg_norm = np.mean(m[112-25:112+25,:])
print 'average 2-norm in middle 50 rows is ', smooth_avg_norm
print('latter / former is %.4f' % (smooth_avg_norm/patchy_avg_norm))
```

Here's a back-of-napkin calculation explaining why the top and bottom vectors don't make much of a difference:
 - The patch vectors are about ~30 smaller than the smooth vectors. 
 - The 4th example looks like it has at most a 16px displacement. So assuming the smooth field is correct, its members will create displacements of ~8px.
 - This means the members of the patchy field create displacements of ~8px/30. I'm using a bilinear sampler, which will take these vectors and return something similar to what the identity vectors would return.
