```python
import sys, h5py, math
import numpy as np
import scipy, scipy.ndimage as img
from data.cv_sampler import Sampler
import matplotlib.pyplot as plt
```

```python
dim = 256  
mip = 2

sampler = Sampler(dim = dim, mip = mip)

def get_chunk():
    chunk = None
    while chunk is None:
        chunk = sampler.random_sample(train=False, two=True)
        counts = dict(zip(*np.unique(chunk, return_counts=True)))
        if 0 in counts:  
            print '  tossing a chunk with a 0'
            chunk = None
    return chunk

chunk = get_chunk()
```

```python
frame = chunk[:,:,0]
target = chunk[:,:,1]
plt.imshow(frame)
plt.show()
plt.imshow(target)
plt.show()
```
