```python
is_test = False   # args.test
mip = 2
final_dim = 256  # dimension of pairs we save                                               
margin = 0.125   # on each side   

from data.cv_sampler import Sampler
big_dim = int(math.ceil(final_dim / (1 - 2*margin)))
sampler = Sampler(source='gs://neuroglancer/pinky40_alignment/prealigned', \
                  mip=mip, dim=big_dim, is_test=is_test)
#chunk = sampler.get_chunk()
m = 256
(x,y,z) = (20505-m*2**2, 31061-m*2**2, 551); d = 2*m*2**2
chunk0 = sampler.chunk_at_global_coords((x,y,z-1),(x+d,y+d,z))
chunk1 = sampler.chunk_at_global_coords((x,y,z),(x+d,y+d,z+1))
chunk2 = sampler.chunk_at_global_coords((x,y,z+1),(x+d,y+d,z+2))
```

```python
dim = 256

frame_dset = np.empty((5, dim, dim))
target_dset = np.empty((5, dim, dim))

for i, (a,b) in enumerate([(1000, 800), (1000, 600), (1200, 600), 
                           (800, 600), (500, 600)]):
    frame_dset[i,:,:] = chunk1.T[a:a+dim,b:b+dim]
    target_dset[i,:,:] = chunk0.T[a:a+dim,b:b+dim]

h5f = h5py.File("../data/crack/natural/" + "simple5.h5", "w")
h5f.create_dataset(u'frame', data=frame_dset)
h5f.create_dataset(u'target', data=target_dset)
h5f.close()
```
